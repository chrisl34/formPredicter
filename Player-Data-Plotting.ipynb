```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('./Shooting Form Data/path_detail.csv')

player_ids = [201935, 202695, 2544, 2594, 201939, 977, 202689, 202681, 201565, 203081]
player_names = ["harden", "kawhi", "LeBron", "Korver", "curry", "kobe", "Kemba", "Kyrie", "drose", "dame"]


for id, name in zip(player_ids, player_names):
    df_player = df[df['pid'] == id]
    cy_norm = (df_player['cy'] - df_player['cy'].min()) / (df_player['cy'].max() - df_player['cy'].min())
    cz_norm = (df_player['cz'] - df_player['cz'].min()) / (df_player['cz'].max() - df_player['cz'].min())

    plt.figure()
    plt.scatter(cy_norm, cz_norm)
    plt.xlabel('Normalized cy')
    plt.ylabel('Normalized cz')
    plt.title(f'{name.capitalize()} (pid={id}) Shooting Form Data')
    plt.show()
```
